```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
dataset_path = 'Map-Crime_Incidents-Previous_Three_Months.csv'
```

```python
SF = pd.read_csv(dataset_path)
SF.head()
```

```python
len(SF)
```

```python
SF['Month'] = SF['Date'].apply(lambda row:int(row[0:2]))
SF['Day'] = SF['Date'].apply(lambda row:int(row[3:5]))
```

```python
SF.head()
```

```python
del SF['IncidntNum']
```

```python
SF.head()
```

```python
SF.drop('Location', axis=1, inplace=True)
SF.head()
#Usar o no implace lo hace definitivo o no
```

```python
contarCategorias = SF['PdDistrict'].value_counts(ascending=True)
contarCategorias.head(1)
```
